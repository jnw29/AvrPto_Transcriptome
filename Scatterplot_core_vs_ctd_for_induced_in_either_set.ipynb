```python
#import library
import numpy as np
import pylab as pl
%matplotlib inline
import scipy.optimize as spo
from scipy import stats
import xlrd as xl
```

```python
#Load book, define sheet as X1, print the number of rows, set n to number of rows
#of sheet 1 column 1, numbers is column,row format.  List is X1a, creates array z.
X = xl.open_workbook('avpfloat.xlsx')
X1 = X.sheet_by_name('NoFormulas')
```

```python
#id test, should be 34730 x 28
m =X1.ncols
n =X1.nrows
print 'cols',m,'rows',n
```

```python
#find the index number of your column - a=0 z=25, aa=26
import string
string.lowercase.index('l')
```

```python
#test and append if either BD (CTD) or CD (core) significantly regulate relative to inactive AvrPto
#pbd-> pvalue BD
#pcd-> pvalue CD
#rbd-> ratio BD
#rcd-> ratio CD
#b-> RPKM b_mean
#c-> RPKM c_mean
#d-> RPKM D_mean
#tbd-> Test for if CTD (BD comparison) meets all criteria
#tcd-> Test for if core (CD comparison) meets all criteria
#BD-> list of BD induction values when logic test is true
#CD-> list of CD induction values when logic test is true

BD = []
CD = []
for i in range (1,n):
    pbd= X1.cell_value(rowx=i, colx=9)
    pcd= X1.cell_value(rowx=i, colx=13)
    rbd= X1.cell_value(rowx=i, colx=8)
    rcd= X1.cell_value(rowx=i, colx=12)
    b= X1.cell_value(rowx=i, colx=7)
    c= X1.cell_value(rowx=i, colx=11)
    d= X1.cell_value(rowx=i, colx=6)
    tbd=0
    tcd=0
    if (pbd<=0.05 and (rbd>=2 or rbd<=0.5) and (b>=3 or d>=3)):
        tbd=1
    if (pcd<=0.05 and (rcd>=2 or rcd<=0.5) and (c>=3 or d>=3)):
        tcd=1
    if tbd==1 or tcd==1:
        BD.append(rbd)
        CD.append(rcd)
        
```

```python
#regression calculation linear, no log transfomration applied
slope, intercept, r_value, p_value, std_err = stats.linregress(BD,CD)
print "slope =", slope
print "intercept =", intercept
print "r-squared =", r_value**2
```

```python
#README: Run this box before running the scatterplot! Intercept and slope numbers (with floating decimal places) are
#used directly in the chart regression calculation on log-transformed set, using base 2. These values are used in the
#next chart.
logBD = np.log2(BD)
logCD = np.log2(CD)
slope, intercept, r_value, p_value, std_err = stats.linregress(logBD,logCD)
print "slope =", slope
print "intercept =", intercept
print "r-squared =", r_value**2
print "2^intercept", 2**intercept
```

```python
import matplotlib.pyplot as plt
#Scatter plot generation
plt.scatter(BD, CD, s=1, alpha=1, color='k', linewidth=None)

#Regression line - log base 2-transformed data's linear regression.
# log(y) = m*log(x)+b is equivalent to y = (x^m)*(2^b) in a log base 2 system.
#README! NOTE: The r^2 value is from the log transformed set of data's linear regression!
z = np.linspace(.07, 6, 100)
plt.plot(z, ((z**slope)*(2**intercept)), '--', color = 'k', linewidth = 2)
plt.figtext(.65, .55, '$y=0.99x^{1.5}$\n$R^2=0.96$', backgroundcolor = 'w', fontsize=16, bbox=dict(facecolor='none', edgecolor='k'))

#Logarithmic scales with grid
plt.xscale('log', basex=2)
plt.yscale('log', basey=2)
plt.grid(True)

#Font size
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   :  14}

plt.rc('font', **font)

axis_font = {'fontname':'Serif', 'size':'14', 'weight':'bold'}
title_font = {'fontname': 'Serif', 'size':'16', 'weight':'bold'}

#set x-axis ticks
xtck = [.0625, .125, .25, .5, 1, 2, 4, 8]
xlbl = ['1/16', '1/8', '1/4', '1/2', '1', '2', '4', '8']
plt.xticks(xtck,xlbl,y=-0.02, rotation=270)

#set y-axis ticks
ytck= [0.015625, 0.03125, .0625, .125, .25, .5, 1, 2, 4, 8, 16]
ylbl= ['1/64', '1/32', '1/16', '1/8', '1/4', '1/2', '1', '2', '4', '8', '16']
plt.yticks(ytck,ylbl, x=-0.02)

#Axis labels

plt.ylabel('CD RPKM / CD$^{-}$/CTD$^{-}$ RPKM', **axis_font)
plt.xlabel('CTD RPKM / CD$^{-}$/CTD$^{-}$ RPKM', **axis_font)

#Reference lines

plt.axhline(y=1, color='k')
plt.axvline(x=1, color='k')

#Plot size and settings
fig = plt.gcf()
fig.set_size_inches(8,8)
fig.set_dpi(300)
plt.tight_layout()
fig.savefig('CoreVsCterm.png',bbox_inches='tight',dpi=300)
plt.show()
```
