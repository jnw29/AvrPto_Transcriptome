```python
#import library
import numpy as np
import pylab as pl
%matplotlib inline
import scipy.stats as stats
import xlrd as xl
import matplotlib.pyplot as plt
import matplotlib.lines as mln
```

```python
#Load book, define sheet as X1, print the number of rows, set n to number of rows
#of sheet 1 column 1, numbers is column,row format.  List is X1a, creates array z.
X = xl.open_workbook('avpfloat.xlsx')
X1 = X.sheet_by_name('NoFormulas')
```

```python
#id test, should be 34730 x 28
m =X1.ncols
n =X1.nrows
print 'cols',m,'rows',n
```

```python
#find the index number of your column - a=0 z=25, aa=26
import string
string.lowercase.index('l')
```

```python
#test and append if AD shows significant regulation. Append ratios of AD (WT), BD(CTD), CD(core)
#THIS SORTS ACCORDING TO AD VALUE! INCREASING. Sorts MD according to AD values.
#pad-> pvalue AD
#rad-> ratio AD
#rbd-> ratio BD
#rcd-> ratio CD
#a-> RPKM b_mean
#d-> RPKM D_mean
#AD-> list of AD induction values when logic test is true
#MD-> list of 
#ADMD-> Array such that ([AD.0,MD.0], [AD.1],[MD.1], ...)

AD = []
BD = []
CD = []
MD = []
for i in range (1,n):
    pad= X1.cell_value(rowx=i, colx=5)
    rad= X1.cell_value(rowx=i, colx=4)
    rbd= X1.cell_value(rowx=i, colx=8)
    rcd= X1.cell_value(rowx=i, colx=12)
    a= X1.cell_value(rowx=i, colx=3)
    d= X1.cell_value(rowx=i, colx=6)
    if (pad<=0.05 and rad<=0.5 and (a>=3 or d>=3)):
        AD.append(rad)
        MD.append(rcd*rbd)
ADMD=zip(AD, MD)
ADMDSorted = sorted(ADMD, key=lambda x: x[0])
AD,MD=zip(*ADMDSorted)  
        
```

```python
#Short demonstration of the inelegant moving-average code used below
IN=range(1,21)
w=5
TEMP = []
zeros=np.zeros((w-1)/2).tolist()
for i in range(((w-1)/2),(len(IN)-((w-1)/2))):
    j = (i-((w-1)/2))
    k = (i+((w+1)/2))
    TEMP.append(sum(IN[j:k])/w)
print 'IN =',IN, 'length =',len(IN)
print 'IN last slice =',IN[j:k]
OUT= zeros+TEMP+zeros
print 'OUT =',OUT,'length =',len(OUT)
```

```python
#create a moving average for each gene where the window size is full
#MA-> Moving average list
#w->window size. MUST BE AN ODD INTEGER.
w = 21
MA = []
zeros=np.zeros((w-1)/2).tolist()
for i in range(((w-1)/2),(len(MD)-((w-1)/2))):
    j = (i-((w-1)/2))
    k = (i+((w+1)/2))
    MA.append(sum(MD[j:k])/w)
MAX=zeros+MA+zeros
```

```python
#linear regression of log base 2 transformed data for the moving average of the two domains combined(Y) plotted against the WT.
#MA-> Moving average list from above. This is missing the first and last 10 data points.
#They are replaced by zeroes for the graphical representation.
#ADS-> AD-Sliced. This is missing the first and last 10 data points after sorting, the same as the moving average.

ADS = AD[10:len(AD)-10]
logADS = np.log2(ADS)
logMA = np.log2(MA)
slope, intercept, r_value, p_value, std_err = stats.linregress(logADS,logMA)
print "slope =", slope
print "intercept =", intercept
print "r-squared =", r_value**2
print "2^intercept", 2**intercept
print "Linear regression of log is y =",slope,"x +",intercept
print "Power function derived from linear regression is y =",2**intercept,"x^",slope
```

```python
#Scatter plot generation
mult = plt.scatter(AD, MD, s=1, alpha=0.3, color='r', linewidth=None)
wild_type = plt.scatter(AD, AD, s=1, alpha=1, color='k', linewidth=None)
movav = plt.plot(AD, MAX, alpha=1, color='b', linewidth=1)

#Logarithmic scales with grid
plt.xscale('log', basex=2)
plt.yscale('log', basey=2)
plt.grid(True)

#Font size
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   :  14}

plt.rc('font', **font)

axis_font = {'fontname':'Serif', 'size':'14', 'weight':'bold'}
title_font = {'fontname': 'Serif', 'size':'16', 'weight':'bold'}

#set x-axis ticks
xtck = [0.00390625, 0.0078125, 0.015625, 0.03125, .0625, .125, .25, .5, 1]
xlbl = ['1/256', '1/128', '1/64', '1/32', '1/16', '1/8', '1/4', '1/2', '1']
plt.xticks(xtck,xlbl,y=-0.02, rotation=270)

#set y-axis ticks
ytck= [0.001953125, 0.00390625, 0.0078125, 0.015625, 0.03125, .0625, .125, .25, .5, 1, 2]
ylbl= ['1/512', '1/256', '1/128', '1/64', '1/32', '1/16', '1/8', '1/4', '1/2', '1', '2']
plt.yticks(ytck,ylbl, x=-0.02)
ax = plt.gca()
ax.set_ylim([0.001953125,2])

#Axis labels

plt.ylabel('Sample RPKM / CD$^{-}$/CTD$^{-}$ RPKM', **axis_font)
plt.xlabel('WT AvrPto RPKM / CD$^{-}$/CTD$^{-}$ RPKM', **axis_font)

#Reference lines

plt.axhline(y=1, color='k')
plt.axvline(x=1, color='k')

#Legend
bline = mln.Line2D([], [], color='blue', marker='_',
                          markersize=1, linewidth = 2)

plt.legend((wild_type, mult, bline),('WT AvrPto', 'CD x CTD', 'Moving average (21 gene window)'),scatterpoints=1,markerscale=4,loc='upper left',ncol=1,fontsize=14)

#Plot size and settings
fig = plt.gcf()
fig.set_size_inches(8,8)
fig.set_dpi(300)
plt.tight_layout()
fig.savefig('Moving Avg Window 21.png',bbox_inches='tight',dpi=300)
plt.show()
```
