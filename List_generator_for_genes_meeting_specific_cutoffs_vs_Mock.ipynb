```python
#import modules
import numpy as np
import xlrd as xl
import xlwt
```

```python
#Load book, define sheet as X1, print the number of rows, set n to number of rows
#of sheet 1 column 1, numbers is column,row format.  List is X1a, creates array z.
#Create new Workbook
X1 = xl.open_workbook('avpfloat.xlsx').sheet_by_name("NoFormulas")
X2 = xl.open_workbook('avpfloat.xlsx').sheet_by_name("NoFormulasBECE")
book=xlwt.Workbook()
```

```python
#ID test, should be 34730 x 28
m =X1.ncols
n =X1.nrows
print n,"x",m
```

```python
#ID test, should be 34730 x 10
m =X2.ncols
n =X2.nrows
print n,"x",m
```

```python
#Test and append up and down regulated genes of each domain
#a-> RPKM A_mean (WT)
#b-> RPKM B_mean (I96A)
#c-> RPKM C_mean (2xA)
#d-> RPKM D_mean (I96A + 2xA)
#e-> RPKM E_mean (Mock)
#pae-> pvalue AD
#pbe-> pvalue BD
#pce-> pvalue CD
#pde-> pvalue DE
#rae-> Ratio AD
#rbe-> Ratio BD
#rce-> Ratio CD
#rde-> Ratio DE
#AvrPtoUp-> List of genes significantly up-regulated in D29E pCPP45::avrPto relative to Mock
#AvrPtoDown-> List of genes significantly down-regulated in D29E pCPP45::avrPto relative to Mock
#CTDUp-> List of genes significantly up-regulated in D29E pCPP45::avrPto I96A relative to D29E Mock
#CTDDown-> List of genes significantly down-regulated in D29E pCPP45::avrPto I96A relative to Mock
#CoreUp-> List of genes significantly up-regulated in D29E pCPP45::avrPto 2xA relative to Mock
#CoreDown-> List of genes significantly down-regulated in D29E pCPP45::avrPto 2xA relative to Mock
#BothUp-> List of genes significantly up-regulated in D29E pCPP45 AvrPto I96A, 2xA relative to Mock
#BothDown-> List of genes significantly down-regulated in D29E pCPP45 AvrPto I96A, 2xA relative to Mock
#Description, Ratio and pValue lists also made - shortened titles.ID = GeneID, D = description
#R = Ratio, P = adjust p

AvrPtoUpID= []
AvrPtoUpD= []
AvrPtoUpR= []
AvrPtoUpP= []
AvrPtoDownID= []
AvrPtoDownD= []
AvrPtoDownR= []
AvrPtoDownP= []
CTDUpID= []
CTDUpD= []
CTDUpR= []
CTDUpP= []
CTDDownID= []
CTDDownD= []
CTDDownR= []
CTDDownP= []
CoreUpID= []
CoreUpD= []
CoreUpR= []
CoreUpP= []
CoreDownID=[]
CoreDownD= []
CoreDownR= []
CoreDownP= []
BothUpID= []
BothUpD= []
BothUpR= []
BothUpP= []
BothDownID= []
BothDownD= []
BothDownR= []
BothDownP= []

AvrPtoUpID.append('GeneID')
AvrPtoUpD.append('Description')
AvrPtoUpR.append('Ratio')
AvrPtoUpP.append('Adjust p')
AvrPtoDownID.append('GeneID')
AvrPtoDownD.append('Description')
AvrPtoDownR.append('Ratio')
AvrPtoDownP.append('Adjust p')
CTDUpID.append('GeneID')
CTDUpD.append('Description')
CTDUpR.append('Ratio')
CTDUpP.append('Adjust p')
CTDDownID.append('GeneID')
CTDDownD.append('Description')
CTDDownR.append('Ratio')
CTDDownP.append('Adjust p')
CoreUpID.append('GeneID')
CoreUpD.append('Description')
CoreUpR.append('Ratio')
CoreUpP.append('Adjust p')
CoreDownID.append('GeneID')
CoreDownD.append('Description')
CoreDownR.append('Ratio')
CoreDownP.append('Adjust p')
BothUpID.append('GeneID')
BothUpD.append('Description')
BothUpR.append('Ratio')
BothUpP.append('Adjust p')
BothDownID.append('GeneID')
BothDownD.append('Description')
BothDownR.append('Ratio')
BothDownP.append('Adjust p')

for i in range (1,n):
    a= X1.cell_value(rowx=i, colx=3)
    b= X2.cell_value(rowx=i, colx=3)
    c= X2.cell_value(rowx=i, colx=7)
    d= X1.cell_value(rowx=i, colx=10)
    e1= X1.cell_value (rowx=i, colx=14)
    e2= X2.cell_value (rowx=i, colx=2)
    pae= X1.cell_value(rowx=i, colx=17)
    pbe= X2.cell_value(rowx=i, colx=5)
    pce= X2.cell_value(rowx=i, colx=9)
    pde= X1.cell_value(rowx=i, colx=21)
    rae= X1.cell_value(rowx=i, colx=16)
    rbe= X2.cell_value(rowx=i, colx=4)
    rce= X2.cell_value(rowx=i, colx=8)
    rde= X1.cell_value(rowx=i, colx=20)
    if (pae<=0.5 and rae>=2 and (a>=3 or e1>=3)):
        AvrPtoUpID.append(X1.cell_value(i,0).encode('utf8'))
        AvrPtoUpD.append(X1.cell_value(i,1).encode('utf8'))
        AvrPtoUpR.append(X1.cell_value(i,16))
        AvrPtoUpP.append(X1.cell_value(i,17))
    if (pae<=0.5 and rae<=0.5 and (a>=3 or e1>=3)):
        AvrPtoDownID.append(X1.cell_value(i,0).encode('utf8'))
        AvrPtoDownD.append(X1.cell_value(i,1).encode('utf8'))
        AvrPtoDownR.append(X1.cell_value(i,16))
        AvrPtoDownP.append(X1.cell_value(i,17))
    if (pbe<=0.5 and rbe>=2 and (b>=3 or e2>=3)):
        CTDUpID.append(X2.cell_value(i,0).encode('utf8'))
        CTDUpD.append(X2.cell_value(i,1).encode('utf8'))
        CTDUpR.append(X2.cell_value(i,4))
        CTDUpP.append(X2.cell_value(i,5))
    if (pbe<=0.5 and rbe<=0.5 and (b>=3 or e2>=3)):
        CTDDownID.append(X2.cell_value(i,0).encode('utf8'))
        CTDDownD.append(X2.cell_value(i,1).encode('utf8'))
        CTDDownR.append(X2.cell_value(i,4))
        CTDDownP.append(X2.cell_value(i,5))
    if (pce<=0.5 and rce>=2 and (c>=3 or e2>=3)):
        CoreUpID.append(X2.cell_value(i,0).encode('utf8'))
        CoreUpD.append(X2.cell_value(i,1).encode('utf8'))
        CoreUpR.append(X2.cell_value(i,8))
        CoreUpP.append(X2.cell_value(i,9))
    if (pce<=0.5 and rce<=0.5 and (c>=3 or e2>=3)):
        CoreDownID.append(X2.cell_value(i,0).encode('utf8'))
        CoreDownD.append(X2.cell_value(i,1).encode('utf8'))
        CoreDownR.append(X2.cell_value(i,8))
        CoreDownP.append(X2.cell_value(i,9))
    if (pde<=0.5 and rde>=2 and (d>=3 or e1>=3)):
        BothUpID.append(X1.cell_value(i,0).encode('utf8'))
        BothUpD.append(X1.cell_value(i,1).encode('utf8'))
        BothUpR.append(X1.cell_value(i,20))
        BothUpP.append(X1.cell_value(i,21))
    if (pde<=0.5 and rde<=0.5 and (d>=3 or e1>=3)):
        BothDownID.append(X1.cell_value(i,0).encode('utf8'))
        BothDownD.append(X1.cell_value(i,1).encode('utf8'))
        BothDownR.append(X1.cell_value(i,20))
        BothDownP.append(X1.cell_value(i,21))

z=len(AvrPtoUpID)
sheet=book.add_sheet('WT up-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,AvrPtoUpID[i].decode('utf8'))
    sheet.write(i,1,AvrPtoUpD[i].decode('utf8'))
    sheet.write(i,2,AvrPtoUpR[i])
    sheet.write(i,3,AvrPtoUpP[i])
z=len(AvrPtoDownID)
sheet=book.add_sheet('WT down-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,AvrPtoDownID[i].decode('utf8'))
    sheet.write(i,1,AvrPtoDownD[i].decode('utf8'))
    sheet.write(i,2,AvrPtoDownR[i])
    sheet.write(i,3,AvrPtoDownP[i])
z=len(CTDUpID)
sheet=book.add_sheet('CTD up-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,CTDUpID[i].decode('utf8'))
    sheet.write(i,1,CTDUpD[i].decode('utf8'))
    sheet.write(i,2,CTDUpR[i])
    sheet.write(i,3,CTDUpP[i])
z=len(CTDDownID)
sheet=book.add_sheet('CTD down-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,CTDDownID[i].decode('utf8'))
    sheet.write(i,1,CTDDownD[i].decode('utf8'))
    sheet.write(i,2,CTDDownR[i])
    sheet.write(i,3,CTDDownP[i])
z=len(CoreUpID)
sheet=book.add_sheet('Core up-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,CoreUpID[i].decode('utf8'))
    sheet.write(i,1,CoreUpD[i].decode('utf8'))
    sheet.write(i,2,CoreUpR[i])
    sheet.write(i,3,CoreUpP[i])
z=len(CoreDownID)
sheet=book.add_sheet('Core down-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,BothDownID[i].decode('utf8'))
    sheet.write(i,1,BothDownD[i].decode('utf8'))
    sheet.write(i,2,BothDownR[i])
    sheet.write(i,3,BothDownP[i])
z=len(BothUpID)
sheet=book.add_sheet('I96A+2xA up-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,BothUpID[i].decode('utf8'))
    sheet.write(i,1,BothUpD[i].decode('utf8'))
    sheet.write(i,2,BothUpR[i])
    sheet.write(i,3,BothUpP[i])
z=len(BothDownID)
sheet=book.add_sheet('I96A+2xA down-regulated vs Mock')
for i in range (0,z):
    sheet.write(i,0,BothDownID[i].decode('utf8'))
    sheet.write(i,1,BothDownD[i].decode('utf8'))
    sheet.write(i,2,BothDownR[i])
    sheet.write(i,3,BothDownP[i])
```
